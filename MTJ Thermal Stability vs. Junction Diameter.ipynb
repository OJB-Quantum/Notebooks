```python
import numpy as np
import matplotlib.pyplot as plt

#  In a FePd perpendicular magnetic tunnel junction with a diameter of 10 nm, junction thickness of 60 nm, and 1 nm tunnel barrier...

# Constants
Ku = 1e6  # uniaxial anisotropy constant (J/m^3)
t = 60e-9  # thickness of the junction (m)
k_B = 1.380649e-23  # Boltzmann constant (J/K)
T = 300  # room temperature (K)

# Junction diameter range from 0 nm to 100 nm
diameters = np.linspace(0, 100e-9, 100)  # in meters

# Calculate the thermal stability factor for each diameter
thermal_stability = Ku * np.pi * (diameters/2)**2 * t / (k_B * T)

# Plotting
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(diameters * 1e9, thermal_stability, color='blue')  # Convert diameters back to nm for plotting, make color changes
plt.xlabel('Junction Diameter (nm)', fontsize = 18)
plt.ylabel('Thermal Stability Factor (Î”)', fontsize = 18)
plt.title('Thermal Stability vs. Junction Diameter', fontsize = 20)
plt.grid(True)
plt.show()
```
