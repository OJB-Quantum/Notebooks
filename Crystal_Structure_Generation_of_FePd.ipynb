```python
from pymatgen.core.lattice import Lattice
from pymatgen.core.structure import Structure
from ase.visualize import view
from ase import Atoms

# Define lattice parameters for Palladium (Pd) - FCC structure
a = 3.89  # Lattice constant for Palladium in Angstroms

# Create the FCC structure of Palladium using pymatgen
lattice = Lattice.cubic(a)
structure = Structure(lattice, ["Pd", "Pd", "Pd", "Pd"],
                      [[0, 0, 0], [0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]])

# Extract element symbols as strings for ASE Atoms
symbols = [str(element) for element in structure.species]

# Convert pymatgen structure to ASE Atoms for visualization
positions = structure.cart_coords
ase_atoms = Atoms(symbols=symbols, positions=positions, cell=lattice.matrix, pbc=[1, 1, 1])

# View the structure
view(ase_atoms, viewer='x3d')
```

```python
from pymatgen.core.lattice import Lattice
from pymatgen.core.structure import Structure
from ase.visualize import view
from ase import Atoms

# Define lattice parameters for FePd - assuming it's similar to Pd
a = 3.89  # Lattice constant in Angstroms, a rough approximation

# Create an ordered FCC structure of FePd using pymatgen
lattice = Lattice.cubic(a)
structure = Structure(lattice, ["Fe", "Pd", "Fe", "Pd"],
                      [[0, 0, 0], [0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]])

# Extract element symbols as strings for ASE Atoms
symbols = [str(element) for element in structure.species]

# Convert pymatgen structure to ASE Atoms for visualization
positions = structure.cart_coords
ase_atoms = Atoms(symbols=symbols, positions=positions, cell=lattice.matrix, pbc=[1, 1, 1])

# View the structure
view(ase_atoms, viewer='x3d')
```

```python
from pymatgen.core.lattice import Lattice
from pymatgen.core.structure import Structure
from ase.visualize import view
from ase import Atoms

# Define lattice parameters for FePd
a = 3.89  # Lattice constant in Angstroms, a rough approximation

# Create an ordered FCC structure of FePd using pymatgen
lattice = Lattice.cubic(a)
structure = Structure(lattice, ["Fe", "Pd", "Fe", "Pd"],
                      [[0, 0, 0], [0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]])

# Create a 3x3x3 supercell
supercell = structure * (3, 3, 3)

# Extract element symbols as strings and positions for ASE Atoms
symbols = [str(element) for element in supercell.species]
positions = supercell.cart_coords

# Convert pymatgen structure to ASE Atoms for visualization
ase_atoms = Atoms(symbols=symbols, positions=positions, cell=supercell.lattice.matrix, pbc=[1, 1, 1])

# View the structure
view(ase_atoms, viewer='x3d')
```
