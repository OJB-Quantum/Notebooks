```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Constants
B = 3.8e15 # cm^-3K^-3/2, Material constant for Si
k = 8.617e-5 # eV/K, Boltzmann constant
Eg_0 = 1.17 # eV, Band gap energy at 0 K for Si
alpha = 4.73e-4 # eV/K, Temperature coefficient of band gap energy
beta = 636 # K, Constant in the band gap energy equation

# Temperature range in Celsius
T_Celsius = np.arange(400, 1101, 1) # from 400 to 1100 Celsius
T_Kelvin = T_Celsius + 273.15 # Convert Celsius to Kelvin

# Band gap energy as a function of temperature
Eg_T = Eg_0 - (alpha * T_Kelvin**2) / (T_Kelvin + beta)

# Intrinsic carrier concentration as a function of temperature
ni_T = B * T_Kelvin**1.5 * np.exp(-Eg_T / (2 * k * T_Kelvin))

# Plot the results
plt.figure(figsize=(10, 6), dpi=300)

plt.plot(T_Celsius, ni_T, color='green',lw=2)
plt.yscale('log') # Use logarithmic scale for better visualization
plt.title('Intrinsic Carrier Concentration of Si from 400°C to 1100°C', fontdict={'fontsize':20})
plt.xlabel('Temperature (°C)', fontdict={'fontsize':14})
plt.ylabel('Intrinsic Carrier Concentration (cm^-3)', fontdict={'fontsize':14})
plt.grid(True)
plt.show()
```
